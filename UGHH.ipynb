```r
library(ggplot2)
```

```r
ggplot(home_combined) +
  geom_bar(aes(x = region, y = sum_Home_Norm_before, fill = "Before"), stat = "identity", position = "identity") +
  geom_bar(aes(x = region, y = sum_Home_Norm_after, fill = "After"), stat = "identity", position = position_dodge(width = 0.8)) +
  labs(title = "Side-by-Side Bar Chart of Home Norm Before and After by Region",
       x = "Region",
       y = "Sum of Home Norm") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust = 1)) +
  scale_fill_manual(values = c("Before" = "#ADD8E6", "After" = "#66c2a5"))
```

```r
library(ggplot2)

# Assuming home_combined is your data frame
home_combined$max_value <- pmax(home_combined$sum_Home_Norm_before, home_combined$sum_Home_Norm_after)

ggplot(home_combined) +
  geom_bar(aes(x = region, y = sum_Home_Norm_before / max_value, fill = "Before"), stat = "identity", position = position_dodge(width = 0.8)) +
  geom_bar(aes(x = region, y = sum_Home_Norm_after / max_value, fill = "After"), stat = "identity", position = position_dodge(width = 0.8)) +
  labs(title = "Clustered Bar Chart of Normalized Home Consumption Before and After by Region",
       x = "Region",
       y = "Relative Normalized Home Consumption") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust = 1)) +
  scale_fill_manual(values = c("Before" = "#ADD8E6", "After" = "#66c2a5"))
```
